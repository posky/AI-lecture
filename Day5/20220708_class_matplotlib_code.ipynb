### 2행 2열의 그래프를 그려보자.
 * 1행 1열: sin() 그래프 
 * 1행 2열: cos() 그래프 (표시 형식:빨간색 사각형)
 * 2행 1열: tan() 그래프 (표시 형식:청록색 점)
 * 2행 2열: arctan() 그래프 (표시 형식:노란색 원)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.0, 2 * np.pi, 0.1)
sin_y = np.sin(x)
cos_y = np.cos(x)
tan_y = np.tan(x)
arctan_y = np.arctan(x)

fig, axs = plt.subplots(2, 2, figsize=(10, 12))
axs[0, 0].plot(x, sin_y)
axs[0, 0].set_title('sinx')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('sinx')

axs[0, 1].plot(x, cos_y, 'rs')
axs[0, 1].set_title('cosx')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('cosx')

axs[1, 0].plot(x, tan_y, 'g.')
axs[1, 0].set_title('tanx')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('tanx')

axs[1, 1].plot(x, arctan_y, 'yo')
axs[1, 1].set_title('arctanx')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('arctanx')

plt.show()
```

```python
x = np.arange(0.0, 2 * np.pi, 0.1)
sin_y = np.sin(x)
cos_y = np.cos(x)
tan_y = np.tan(x)
arctan_y = np.arctan(x)

plt.figure(figsize=(10, 12))

plt.subplot(2, 2, 1)
plt.plot(x, sin_y, label='y = sinx')
plt.title('sinx graph')
plt.xlabel('x')
plt.ylabel('sinx')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x, cos_y, 'rs', label='y = cosx')
plt.title('cosx graph')
plt.xlabel('x')
plt.ylabel('cosx')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x, tan_y, 'g.', label='y = tanx')
plt.title('tanx graph')
plt.xlabel('x')
plt.ylabel('tanx')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x, arctan_y, 'yo', label='y = arctanx')
plt.title('arctanx graph')
plt.xlabel('x')
plt.ylabel('arctanx')
plt.legend()

plt.show()
```
