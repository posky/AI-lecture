```python
loc = folium.Map(location=[35.13375472678323, 126.87521313035022], zoom_start=14)

folium.Marker(
    [35.13375472678323, 126.87521313035022],
    popup='Wold Cup Stadium',
    icon=folium.Icon(color='purple', icon='info-sign')
).add_to(loc)
folium.CircleMarker(
    location=[35.140298169188, 126.85840336856398],
    radius=100,
    color='#ffffgg',
    fill_color='#fffggg',
    popup='Noodle'
).add_to(loc)

loc
```
