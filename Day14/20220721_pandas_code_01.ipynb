```python
import pandas as pd

from pandas import Series, DataFrame
```

```python
dat = {
    'col1': [1, 2, 3, 4],
    'col2': [10, 20, 30, 40],
    'col3': ['A', 'B', 'C', 'D']
}

df = DataFrame(dat)
df
```

```python
team_score = {
    "toto":[1500,3000,5000,7000,5500],
    "apple":[4000,5000,6000,5500,4500],
    "gildong":[2000,2500,3000,4000,3000],
    "catanddog":[7000,5000,3000,5000,4000]
}

team_df = DataFrame(team_score)
team_df
```

```python
date = ['22-05-01','22-05-02', '22-05-03', '22-05-04', '22-05-05']
team_df = DataFrame(
    team_score,
    columns=['catanddog', 'toto', 'apple', 'gildong'],
    index=date
)
team_df
```

* iloc: 인덱스 (숫자) 접근
* loc: 인덱스

```python
team_df.loc['22-05-02':'22-05-03', 'toto':'apple']
```

```python
team_df.loc['22-05-03':'22-05-05', ['toto', 'gildong']]
```

```python
team_df.iloc[2:][['toto', 'gildong']]
```

```python
team_df.iloc[2:, [1, 3]]
```

```python
team_df.loc['22-05-05', 'gildong'] = 7000
team_df
```

```python
team_df.loc[team_df.sum(axis=1) >= 16000]
```

```python
team_df.loc[team_df['gildong'] <= 3000]
```

```python
team_df.loc[(team_df['gildong'] <= 3000) & (team_df['toto'] >= 3000)]
```

```python
team_df['row_sum'] = team_df.sum(axis=1)
team_df
```

```python
team_df.sum()
print(type(team_df.sum()))
```

```python
team_df.loc['sum'] = team_df.sum()
team_df
```

```python
team_df['mean'] = team_df.mean(axis=1)
team_df
```

```python
team_df['std'] = team_df.std(axis=1)
team_df
```

```python
team_df.drop(['mean', 'std'], axis=1, inplace=True)
team_df
```

```python
team_df.drop(['22-05-03', '22-05-04'], axis=0, inplace=True)
team_df
```
