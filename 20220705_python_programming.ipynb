```python
class Cal:
    result = 0
    
    def plus(self, num):
        self.result += num
        return self.result
        
    def minus(self, num):
        self.result -= num
        return self.result
    
    def divide(self, num):
        self.result /= num
        return self.result
```

```python
c1 = Cal()
c2 = Cal()
c3 = Cal()
```

```python
c3.divide(0)
```

```python
class Cal_change(Cal):
    def divide(self, num):
        if num == 0:
            print("0으로 나눌수 없습니다.")
        else:
            self.result /= num
        return self.result
    
    def mul(self, num):
        self.result *= mul
        return self.result
        
    def zero(self):
        self.result = 0
        return self.result
```

```python
c1 = Cal_change()
c2 = Cal_change()
c3 = Cal_change()
c4 = Cal_change()
c5 = Cal_change()
```

```python
print(c1.plus(5))
print(c1.divide(5))
print(c1.zero())
```

```python
# np.linspace(start, end, number)
x = np.linspace(0, 14, 100) # 0부터 14까지 100개
y1 = np.sin(x)              # y1 = sin(x)
y2 = 2 * np.sin(x + 0.5)    # y2 = 2sin(x + 0.5)
y3 = 4 * np.sin(x + 1.0)    # y3 = 4sin(x + 1.0)

plt.figure(figsize = (10, 6))   # figsize (10 x 6) inch
plt.plot(x, y1)                 # graph
```

### seaborn 시각화
 * boxplot : 상자 그림
 * lmplot : 산정도와 회귀 직선
 * barplot : 막대 그래프 (값의 평균)
 * countplot : 막대 그래프 (빈도)
 * scatterplot : 도 변수간의 상관관계 찍기
 * lineplot : 시계열에 따른 데이터 추이
 * histoplot : 히스토그램
 * heatmap : 히트맵
