```python
train['var3'].replace(-999999, 2, inplace=True)
```

```python
train.loc[train['var3'] == -999999, :]
```

```python
train.drop('ID', axis=1, inplace=True)
train.head()
```

```python
X = train.iloc[:, :-1]
y = train['TARGET']

X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    stratify=y,
    test_size=0.2,
    random_state=0
)

X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
print('train label distribution:\n{}'.format(y_train.value_counts() / y_train.count()))
print('test label distribution:\n{}'.format(y_test.value_counts() / y_test.count()))
```

```python
%%time

from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score

xgb_model = XGBClassifier(n_estimators=500, random_state=156)
xgb_model.fit(
    X_train, y_train,
    early_stopping_rounds=100,
    eval_metric='auc',
    eval_set=[(X_train, y_train), (X_test, y_test)]
)
```

```python
pred_prob = xgb_model.predict_proba(X_test)[:, 1]
pred_prob
```

```python
xgb_roc_score = roc_auc_score(y_test, pred_prob, average='macro')
print('ROC AUC: {:.4f}'.format(xgb_roc_score))
```

```python
pred_01 = pred_prob > 0.1
pred_01
```

```python
xgb_roc_score = roc_auc_score(y_test, pred_01, average='macro')
print('ROC AUC: {:.4f}'.format(xgb_roc_score))
```

```python
%%time

from sklearn.model_selection import GridSearchCV

xgb_model1 = XGBClassifier(n_eistmators=100, use_label_encoder=False)
params = {
    'max_depth': [5, 7], 
    'min_child_weight': [1, 3],
    'colsample_bytree': [0.5, 0.75]
}

gridcv = GridSearchCV(xgb_model1, param_grid=params, cv=3)
gridcv.fit(
    X_train, y_train,
    early_stopping_rounds=30,
    eval_metric='auc',
    eval_set=[(X_train, y_train), (X_test, y_test)]
)
```

```python
print('GridSearchCV best parameter:', gridcv.best_params_)

pred_prob = gridcv.predict_proba(X_test)[:, 1]

xgb_roc_score = roc_auc_score(y_test, pred_prob, average='macro')
print('ROC AUC: {:4f}'.format(xgb_roc_score))
```
