* 딥러닝 - Deep Neural Network
    * 입력층 (Input Layer) - 은닉층 (Hidden Layer) - 출력층 (Output Layer)

* 데이터에 의해서 결정되는 층이 2개가 있다.
    * 입력층 - 
    * 출력층
        * 연속형 예측하기 (출력층 뉴런이 하나다.)
        * 범주형 (MNIST, Paddy docter) 예측
            * 출력층은 범주형 개수만큼 뉴런을 만든다.
            * 활성화 함수가 softmax가 된다.
                * softmax 활성화 함수는 값을 전부 더하면 1이다.
                * 출력층의 각 뉴런에서 마지막에 출력되는 값은 확률적으로 이 값이 나올 확률.
* 활성화 함수
    * 종류
        * sigmoid: 0~1
        * relu: 음수로 값이 들어오면 0, 양수는 그대로
    * 활성화 함수가 있는 것과 없는 것의 차이
        * 선형문제만 풀 수 있는 것이 아니라, 비선형 문제까지 해결 가능.
* optimizer란? 손실함수를 최저값으로 만들기 위해서, 가중치 조절하는 함수
    * adam
    * rmsprop
    * sgd
    * gd
* loss 함수란? 마지막 결과값과 실제 가진 값을 오차를 구하는 함수
    * mse (mean squared error)
    * categorical crossentropy
    * binary crossentropy

* CNN이 뭘까요?
    * 이미지 분야에 있어 뛰어난 성능을 낼 수 있다.
    * FCL의 이 한계를 극복하기 위해서 CNN이 적용
        * 공간정보 손실을 극복. 공간 정보의 특징을 살릴 수 있다.
        * 
    * Conv - Pooling 계층과 FCL로 연결 (다 그런건 아님)
        * Conv (컨볼루션) - 이미지의 값을 요약
        * Pooling - 이미지를 다운 샘플링
    * Stride - 필터가 움직이는 크기
    * Filter
        * 가중치를 가지고 있다.
        * 이미지읖 특징을 각 필터가 갖게 되고, 찾을 수 있다.
    * 특징맵(feature map)
        * 필터외 합성곱 연산을 통해서 만들어지는 결과물
    * Pooling에는 가중치가 있을까?
        * Max Pooling
        * Average Pooling
        * Min Pooling
        * 24 x 24 x 16 => 2 x 2 stride 2
            * 12 x 12 x 16
