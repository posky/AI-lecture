```python
from pandas import DataFrame

team_score = {
    "toto": [1500, 3000, 5000, 7000, 5500],
    "apple": [4000, 5000, 6000, 5500, 4500],
    "gildong": [2000, 2500, 3000, 4000, 3000],
    "catanddog": [7000, 5000, 3000, 5000, 4000],
    'banana': [5000, 4000, 3000, 3000, 4000]
}

team_df = DataFrame(team_score)
cols = team_df.columns
print(cols)
team_df
```

```python
team_df['row_sum'] = team_df[cols].sum(axis=1)
team_df
```

```python
team_df['row_mean'] = team_df[cols].mean(axis=1)
team_df
```

```python
team_df['number'] = team_df[cols].count(axis=1)
team_df
```

```python
team_df['new_avg'] = team_df['row_sum'] / len(team_df)
team_df
```
